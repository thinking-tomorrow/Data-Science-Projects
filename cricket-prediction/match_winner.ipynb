```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('matches.csv')
```

```python
df.head()
```

```python
df.drop(['id', 'dl_applied', 'win_by_runs', 'win_by_wickets' ,'player_of_match', 'umpire1', 'umpire2', 'umpire3', 'date', 'city'], axis=1, inplace=True)
```

```python
df['season']=df['season'].astype('int')
```

```python
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['team1'].unique()
```

```python
teams=['Sunrisers Hyderabad', 'Mumbai Indians', 'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Kings XI Punjab','Chennai Super Kings', 
       'Rajasthan Royals', 'Delhi Capitals']
df=df[(df['team1'].isin(teams))&(df['team2'].isin(teams))]
```

```python
df.replace('Delhi Daredevils', 'Delhi Capitals', inplace=True)
```

```python
teams=list(enumerate(df['team1'].unique(), start=1))
map_list={value:key for key, value in teams}
```

```python
map_list
```

```python
df[['team1', 'team2', 'toss_winner', 'winner']]=df[['team1', 'team2', 'toss_winner', 'winner']].apply(lambda x:x.map(map_list), axis=1)
```

```python
df=pd.get_dummies(df, columns=['toss_decision', 'result', 'venue'], drop_first=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
X=df.drop('winner', axis=1)
y=df['winner']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2)
```

```python
log_reg=LogisticRegression()
log_reg.fit(X_train, y_train)
log_reg.score(X_test, y_test)
```

```python
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt.score(X_test, y_test)
```
