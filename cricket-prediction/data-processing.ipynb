```python
from sqlalchemy import create_engine
engine = create_engine('mysql+pymysql://root:@localhost/cricket_prediction', echo=False)
```

```python
df.to_sql('matches', engine)
```

```python
def get_match_winner_list():
    sql="SELECT winner, COUNT(*) AS 'num' FROM matches GROUP BY winner ORDER BY COUNT(*) DESC"
    with engine.connect() as connection:
        result = connection.execute(sql)
        return result.fetchall()
```

```python
x = df['team1'].unique()
```

```python
teams = list(x)
```

```python
def get_match_list():
    for i in teams:
        sql=f"SELECT COUNT(*) FROM matches WHERE team1='{i}' OR team2='{i}'"
        with engine.connect() as connection:
            result = connection.exoecute(sql)
            return result.fetchone()
```
